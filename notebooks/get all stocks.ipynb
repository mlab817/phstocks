```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import json
```

```python
url = 'https://frames.pse.com.ph/indices/all'

allStocksPage = requests.get(url)
soup = BeautifulSoup(allStocksPage.text, 'html.parser')
```

```python
table = soup.find('table', id='data')
trs = table.find_all('tr')

allStocks = []

for tr in trs[1:]:
    tds = tr.find_all('td')
    allStocks.append({
        'name': tds[1].text,
        'url': tds[2].find('a', href=True)['href'],
        'symbol': tds[2].text,
        'last_traded_date': tds[3].text,
        'last_traded_price': tds[4].text,
        'outstanding_shares': tds[5].text
    })
    
allStocks
```

```python
df = pd.DataFrame(allStocks)
df.head()
```

```python
df.last_traded_price = df.last_traded_price.str.replace(',','').astype(float)
```

```python
df.outstanding_shares = df.outstanding_shares.str.replace(',','').astype(int)
```

```python
df.to_csv('stocks.csv', encoding='utf-8')
```
